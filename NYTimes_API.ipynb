```python
import requests
import pandas as pd
from pandas.io.json import json_normalize
```

```python
# NYTimes Books API
# See: https://developer.nytimes.com/books_api.json#/Console/GET/lists/best-sellers/history.json
API_KEY = '61b5a1a0f9a445c9ad40b264b1f2f995'
url = "https://api.nytimes.com/svc/books/v3/lists/best-sellers/history.json?"
```

```python
# Make the API request
request = requests.get(url, params={'api-key': API_KEY})
data = request.json()
```

```python
# JSON result
data
```

```python
# Flatten JSON so it's more usable in a dataframe
meta = ['author', 'title', 'publisher', 'description']
df = json_normalize(data['results'], 'ranks_history', meta=meta)
```
